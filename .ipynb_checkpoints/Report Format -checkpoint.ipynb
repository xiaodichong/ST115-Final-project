### 3. Data Description<!-- {#section-3} -->

1. Student ratings of varying aspects of their educational experience by subject of study
 - Student ratings were acquired from the National Student Survey, where results were filtered by provider, which was chosen as LSE.
 
 - Once acquired, the data was filtered to only include undergraduate, full-time student data, to avoid overcomplicating the analysis and adding unnecessary noise to our analysis. 
 
 - Student ratings were obtained upon 
   - Quality of Teaching
   - Learning Opportunities
   - Assessment and Feedback
   - Academic Support
   - Organisation and Management
   - Learning Resources
   - Student Voice
 - There was no missing data for any of the entries. However, due to the way the subjects were sorted, we used the most precise definition of subject to identify each subjet. For example, we did not use data for 'Mathematical Sciences', and instead used data for 'Mathematics' and 'Statistics' seperately.

2. Reddit posts and comments
- Reddit posts and their comments were obtained from the "r/6thform" subreddit using the Reddit API. 
- Posts collected were posts related to or containing keywords such as the abbreviated university names e.g. LSE
- Data from the Reddit posts was ot only collected about LSE but also UCL, Imperial College London, Kings College London and Oxbridge universities. 
- Additional information about the posts were also obtained such as the flair associated with the post and the date that each comment and post was posted to the subreddit
